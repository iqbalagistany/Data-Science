```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("ANTM.csv")
```

```python
df.head()
```

Rubah settingan tanggal

```python
df['date'] = pd.to_datetime(df.date)
df.head()
```

Hapus kolom yang tidak perlu

```python
df.drop(df.iloc[:, 7:], inplace = True, axis = 1)
```

```python
df.head()
```

**SIMPLE MOVING AVERAGE**

```python
sma10 = df["Simple Moving Average"] = df["close"].rolling(window = 10).mean()
```

```python
df.head(15)
```

```python
df.plot.line(x="date", y="Simple Moving Average")
```

```python
df.plot.line(x="date", y=["close", "Simple Moving Average"], figsize=(20,6), title = "MOVING AVERAGE")
```

**Weighted Moving Average**

```python
#Buat array int untuk dijadikan weight
weights = np.arange(1,11) 
weights
```

```python
wma10 = df['close'].rolling(10).apply(lambda prices: np.dot(prices, weights)/weights.sum(), raw=True)
```

```python
wma10.head(20)
```

```python
df['Weighted Moving Average'] = np.round(wma10, decimals=2)
```

```python
df.head(20)
```

```python
df.plot.line(x="date", y=["close", "Simple Moving Average", "Weighted Moving Average"], figsize=(20,6), title = "MOVING AVERAGE")
```

**Exponential Moving Average**

```python
df.head(20)
```

```python
ema10 = df['close'].ewm(span=10).mean()
ema10.head(10)
```

```python
df['Exponential Moving Average'] = np.round(ema10, decimals=2)
```

```python
df.head(20)
```

```python
df[["date",'close', 'Simple Moving Average', 'Weighted Moving Average', "Exponential Moving Average" ]].head(20)
```

Agar dimulai dari nilai ke-10

```python
modPrice = df['close'].copy()
sma10 = df["close"].rolling(window = 10).mean()
modPrice.iloc[0:10] = sma10[0:10]
modPrice.head(20)
```

```python
ema10alt = modPrice.ewm(span=10, adjust=False).mean()
```

```python
df['Exponential Moving Average'] = np.round(ema10alt, decimals=2)
df[['date', 'close', 'Simple Moving Average', 'Weighted Moving Average', "Exponential Moving Average"]].head(20)
```

```python
df.plot.line(x="date", y=["close", "Simple Moving Average", "Weighted Moving Average", "Exponential Moving Average"], figsize=(20,6), title = "MOVING AVERAGE")
```
